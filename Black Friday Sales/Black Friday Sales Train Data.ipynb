```python
from sklearn.model_selection import train_test_split
x = df.drop("Purchase",axis = 1)
y = df["Purchase"]
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_pred = LR.predict(x_test)

print(y_test.values[:5])
print(y_pred[:5])
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
mse1 = mean_squared_error(y_test,y_pred)
mae1 = mean_absolute_error(y_test,y_pred)
r2_score1 = r2_score(y_test,y_pred)
```

```python
print(f"mse: {mse1}")
print(f"mae : {mae1}")
print(f"r2score : {r2_score1}")
```

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor()
DT.fit(x_train,y_train)
```

```python
y_pred = DT.predict(x_test)

print(y_test.values[:5])
print(y_pred[:5])
```

```python
mse2 = mean_squared_error(y_test,y_pred)
mae2 = mean_absolute_error(y_test,y_pred)
r2_score2 = r2_score(y_test,y_pred)
```

```python
print(f"mse: {mse2}")
print(f"mae : {mae2}")
print(f"r2score : {r2_score2}")
```

```python
from sklearn.neighbors import KNeighborsRegressor
KNN = KNeighborsRegressor()
KNN.fit(x_train,y_train)
```

```python
y_pred = KNN.predict(x_test)

print(y_test.values[:5])
print(y_pred[:5])
```

```python
mse3 = mean_squared_error(y_test,y_pred)
mae3 = mean_absolute_error(y_test,y_pred)
r2_score3 = r2_score(y_test,y_pred)
```

```python
print(f"mse: {mse3}")
print(f"mae : {mae3}")
print(f"r2score : {r2_score3}")
```

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()
RF.fit(x_train,y_train)
```

```python
y_pred = RF.predict(x_test)

print(y_test.values[:5])
print(y_pred[:5])
```

```python
mse4 = mean_squared_error(y_test,y_pred)
mae4 = mean_absolute_error(y_test,y_pred)
r2_score4 = r2_score(y_test,y_pred)
```

```python
print(f"mse: {mse4}")
print(f"mae : {mae4}")
print(f"r2score : {r2_score4}")
```

```python
dict1 = {
    "Mean Square Error" : [mse1,mse2,mse3,mse4],
    "Mean Absolute Error" : [mae1,mae2,mae3,mae4],
    "R2 Score" : [r2_score1,r2_score2,r2_score3,r2_score4]}
pd.DataFrame(dict1,index=["Linear Regression","Decision Tree","KNN","Random Forest"])
```
